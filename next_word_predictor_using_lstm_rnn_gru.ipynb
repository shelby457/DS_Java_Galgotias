```python
text=[]
with open('/content/next_word_pred.txt', 'r', encoding = 'utf-8') as f:
  while (f. readline()):
     text.append (f.readline())
```

```python
input_text=[]
for i in range(0,len(text),2):
  txt=''
  txt=txt+text[i].replace('\t',' ').replace('\n','')
  input_text.append(txt)
```

```python
text
```

```python
input_text
```

```python
def full_stop(text):
  return text.replace(".","").replace("?","").lower()
```

```python
new_input=[]
for i in input_text:
  new_input.append(full_stop(i))
```

```python
new_input[9]
```

```python
import tensorflow as tf
import keras
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts(new_input)
```

```python
number_of_words=len(tokenizer.word_index)-1
```

```python
tokenizer.word_index
```

```python
number_of_words
```

```python
input_sequences=[]
for sentece in input_text:
  tokenized_sentence=tokenizer.texts_to_sequences([sentece])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len=max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
padded_input_sequences=pad_sequences(input_sequences,maxlen=max_len,padding='pre')
```

```python
padded_input_sequences
```

```python
X=padded_input_sequences[:,:-1]
y=padded_input_sequences[:,-1]
```
